```python
query = '''
WITH 
month AS (
    SELECT DISTINCT customer_id, order_month FROM orders ORDER BY customer_id
),
userid AS (
    SELECT DISTINCT customer_id FROM orders
),
period AS (
    SELECT DISTINCT order_month FROM month
),
cross_joined AS (
    SELECT * FROM userid CROSS JOIN period
),
joined AS (
    SELECT c.customer_id, c.order_month AS sort_m, m.order_month AS active_m
    FROM cross_joined c
    LEFT JOIN month m ON c.customer_id = m.customer_id AND c.order_month = m.order_month
),
record AS (
    SELECT *,
    LEAD(active_m) OVER (PARTITION BY customer_id ORDER BY sort_m) AS next_m
    FROM joined
),
filter AS (
    SELECT customer_id, sort_m, active_m, next_m FROM record
    WHERE active_m IS NOT NULL
    ORDER BY customer_id, active_m
),
active AS (
    SELECT sort_m, COUNT(DISTINCT(customer_id)) AS active 
    FROM filter
    WHERE active_m IS NOT NULL
    GROUP BY sort_m
    ORDER BY sort_m
),
churned AS (
    SELECT
        active_m, COUNT(DISTINCT(customer_id)) AS churned
    FROM filter
    WHERE next_m IS NULL
    GROUP BY active_m
    ORDER BY active_m
)
SELECT
    a.sort_m AS month,
    COALESCE(c.churned, 0) AS churned_customers,
    COALESCE(a.active, 0) AS total_active_customers,
    CASE
        WHEN COALESCE(a.active, 0) = 0 THEN 0
        ELSE COALESCE(c.churned, 0) * 100.0 / COALESCE(a.active, 0)
    END AS churn_rate
FROM active a
LEFT JOIN churned c
ON a.sort_m = c.active_m
ORDER BY month
'''

result = conn.execute(query).df()

print(result)
```

```python
query = '''
WITH 
quarters AS (
    SELECT DISTINCT
        customer_id,
        CASE
            WHEN EXTRACT(MONTH FROM order_purchase_timestamp) BETWEEN 1 AND 3 THEN CONCAT(EXTRACT(YEAR FROM order_purchase_timestamp), '-Q1')
            WHEN EXTRACT(MONTH FROM order_purchase_timestamp) BETWEEN 4 AND 6 THEN CONCAT(EXTRACT(YEAR FROM order_purchase_timestamp), '-Q2')
            WHEN EXTRACT(MONTH FROM order_purchase_timestamp) BETWEEN 7 AND 9 THEN CONCAT(EXTRACT(YEAR FROM order_purchase_timestamp), '-Q3')
            ELSE CONCAT(EXTRACT(YEAR FROM order_purchase_timestamp), '-Q4')
        END AS quarter
    FROM orders
    ORDER BY customer_id
),
userid AS (
    SELECT DISTINCT 
        customer_id
    FROM orders
),
period AS (
    SELECT DISTINCT
        quarter
    FROM Quarters
),
cross_joined AS (
    SELECT * FROM userid CROSS JOIN period
),
joined AS (
    SELECT 
        c.customer_id, 
        c.quarter AS sort_q, 
        q.quarter AS active_q 
    FROM cross_joined c
    LEFT JOIN quarters q ON c.customer_id = q.customer_id 
    AND c.quarter = q.quarter
),
record AS (
    SELECT 
        *,
        LEAD(active_q) OVER (PARTITION BY customer_id ORDER BY sort_q) AS next_q
    FROM joined
),
filter AS (
    SELECT 
        customer_id, sort_q, active_q, next_q 
    FROM record
    WHERE active_q IS NOT NULL
    ORDER BY customer_id, active_q
),
active AS (
    SELECT 
        sort_q, COUNT(DISTINCT(customer_id)) AS active 
    FROM filter
    WHERE active_q IS NOT NULL
    GROUP BY sort_q
    ORDER BY sort_q
),
churned AS (
    SELECT
        active_q, COUNT(DISTINCT(customer_id)) AS churned
    FROM filter
    WHERE next_q IS NULL
    GROUP BY active_q
    ORDER BY active_q
)
SELECT
    a.sort_q AS quarter,
    COALESCE(c.churned, 0) AS churned_customers,
    COALESCE(a.active, 0) AS total_active_customers,
    CASE
        WHEN COALESCE(a.active, 0) = 0 THEN 0
        ELSE COALESCE(c.churned, 0) * 100.0 / COALESCE(a.active, 0)
    END AS churn_rate
FROM active a
LEFT JOIN churned c
ON a.sort_q = c.active_q
ORDER BY quarter
'''

result = conn.execute(query).df()
print(result)
```

```python
query = '''
WITH 
year AS (
    SELECT DISTINCT customer_id, order_year FROM orders ORDER BY customer_id
),
userid AS (
    SELECT DISTINCT customer_id FROM orders
),
period AS (
    SELECT DISTINCT order_year FROM year
),
cross_joined AS (
    SELECT * FROM userid CROSS JOIN period
),
joined AS (
    SELECT c.customer_id, c.order_year AS sort_y, y.order_year AS active_y
    FROM cross_joined c
    LEFT JOIN year y ON c.customer_id = y.customer_id AND c.order_year = y.order_year
),
record AS (
    SELECT *,
    LEAD(active_y) OVER (PARTITION BY customer_id ORDER BY sort_y) AS next_y
    FROM joined
),
filter AS (
    SELECT customer_id, sort_y, active_y, next_y FROM record
    WHERE active_y IS NOT NULL
    ORDER BY customer_id, active_y
),
active AS (
    SELECT sort_y, COUNT(DISTINCT(customer_id)) AS active 
    FROM filter
    WHERE active_y IS NOT NULL
    GROUP BY sort_y
    ORDER BY sort_y
),
churned AS (
    SELECT
        active_y, COUNT(DISTINCT(customer_id)) AS churned
    FROM filter
    WHERE next_y IS NULL
    GROUP BY active_y
    ORDER BY active_y
)
SELECT
    a.sort_y AS year,
    COALESCE(c.churned, 0) AS churned_customers,
    COALESCE(a.active, 0) AS total_active_customers,
    CASE
        WHEN COALESCE(a.active, 0) = 0 THEN 0
        ELSE COALESCE(c.churned, 0) * 100.0 / COALESCE(a.active, 0)
    END AS churn_rate
FROM active a
LEFT JOIN churned c
ON a.sort_y = c.active_y
ORDER BY year
'''

result = conn.execute(query).df()

print(result)
```
